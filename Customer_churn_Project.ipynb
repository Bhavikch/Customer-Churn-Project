```python
#simple linear regression model - 1 Independent variable
#Multiple linear regression -  More than 1 Independent variable

'''
C) Linear Regression:
a. Build a simple linear model where dependent variable is ‘MonthlyCharges’ and independent
variable is ‘tenure’
i. Divide the dataset into train and test sets in 70:30 ratio.
ii. Build the model on train set and predict the values on test set
iii. After predicting the values, find the root mean square error
iv. Find out the error in prediction & store the result in ‘error’
v. Find the root mean square error

'''


# Import libraries

from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
# x=Independent variable(tenure)
#y=Dependent variable(Monthly charges)
```

```python
x=df.loc[ : ,["tenure"]].values

y=df.loc[:,["MonthlyCharges"]].values
```

```python
# Now we have to divide our data into training and testing

'''
x_train= Question for learning
y_train= Answers for learning

x_test= Exam question !
y_test= Exam Answers - You have to write it
'''

#Random_state = Background pattern for training
```

```python
x_train ,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression

# Name your model

my_model =LinearRegression()
```

```python
# Now i will train my model with training question and training answers

my_model.fit(x_train,y_train)
```

```python
# Now i will test my model by passing EXAM Questions

y_pred=my_model.predict(x_test)
```

```python
# Now i will compare the Actual answer od Exam question with the answer given by my machine

# Compare Actual correct ans with Predicted ans by machine

# I will use MSE to find how far my predicted ans are from Actual ans
```

```python
from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y_test,y_pred)

# Comparing actual value with predicted value

mse
```

```python
#RMSE- Root mean squared Error

rmse=np.sqrt(mse)

rmse
# It will give output in the same unit as original data
```
